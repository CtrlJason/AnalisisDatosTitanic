### Separacion de los nombres
Se utiliza ***split*** para separar cada valor y guardarlos en un arreglo

```python
df['Name'].str.split(',')
```

### Convertimos todos los nombre a minusculas
para ello se utiliza el metodo ***str*** para trabajar los string con python y luego el metodo ***lower*** para convertir todo a minusculas

```python
df['Name'] = df['Name'].str.lower()
```

```python
df
```

### Creacion de columnas: FirstName y LastName
Se crean las columnas en el ***DataFrame*** separando los datos que necesitamos

***Creamos la columna "FirstName" para almacenar los primeros nombres***

```python
df['FirstName'] = df['Name'].str.split(',').str.get(0)

df
```

***Separamos los titulos de "Name" (nombre completo con titulo)***

```python
df['Name'].str.split(',').str.get(1).str.split('.').str.get(0) # Con str.split('.') Separamos los titulos ya que estos terminan con .
```

***Creamos la columna "Status" para almacenar los titulos***

```python
df['Status'] = df['Name'].str.split(',').str.get(1).str.split('.').str.get(0)
df
```

***Creamos la columna "LastName" para almacenar los apellidos***

```python
df['LastName'] = df['Name'].str.split(',').str.get(1).str.split('.').str.get(1)
df
```
